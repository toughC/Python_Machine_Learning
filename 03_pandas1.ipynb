##### 리스트 데이터를 이용한 DataFrame 생성

```python
import pandas as pd

lstdata=[[1, "홍길동", "남" ],
         [2, "김파이", "남", 35],
         [3, '박데이', "여", 55],
         [4, '이분석', '여', 60],
         [5, '박판다', '남', 30]
         ]
         
# 행/열 인덱스를 기본값으로 부여(숫자)
# pandas.DataFrame(2차원 list data, columns=[필드명, ....], index=[행 인덱스 지정])
df=pd.DataFrame(lstdata)  
df
```

```python
df=pd.DataFrame(lstdata, columns=['번호', '이름', '성별', '나이'])  
df
```

##### 딕셔너리 데이터를 이용한 DataFrame 생성

```python
dictdata={"번호":[1,2,3,4,5],
          "이름":["AAA", "BBB", "CCC", "DDD", "EEE"],
          "성별":["남", "남", "여", "여", "남"],
          "나이":[50,40,30,35,40]
          }

df=pd.DataFrame(dictdata)
df
```

##### 행/열 데이터 선택(p21~)
- df.loc[행, 열]  : index, columns 값을 이용한 출력
- df.iloc[행위치, 열위치] : 행, 열의 위치값
- df.sort_index(): index 기준 정렬
- df.sort_values(기준열) : 값을 기준으로 정렬

```python
# 행/열 데이터 선택(.loc[])
df.loc[2]   # 행 index값이 2인 행 출력
df.loc[:,["이름"]] # 이름 열 전체 출력 => df['이름']
```

```python
# 다중 행/열 선택(loc[])
display(df.loc[1:3])    # index가 1~3인 데이터 행 전체 출력
display(df.loc[:, ['이름', '나이']]) # 이름. 나이 열만 출력
print("========================================")
display(df.loc[1:3, ['이름', '나이']])  # index가 1~3 사이의 데이터에서 '이름', '나이' 열 출력
```

```python
# 행 출력(iloc[])
df.iloc[1:3]   # 행의 1위치부터 3전 위치까지 출력
```

```python
# 열 출력(iloc[])
df.iloc[:, [1, 3]]  # 1, 3번 위치값 출력, df.iloc[:, [1, -1]]
```

```python
df.iloc[1:3, 1:3]
```

```python
# 열 순서 변경후 출력
df.loc[:, ['이름', '나이', '성별']]   # df[['이름', '나이', '성별']]
df.iloc[:, [1, -1, 2]]
```

```python
# 특정 원소 선택 및 값 변경
df.loc[0, '나이']
df.loc[0, ['나이']]=48
df
```

```python
df.iloc[0, [-1]]
df.iloc[0, [-1]]=50
df
```

```python
# 데이터 정렬(값)
df.sort_values(['나이'], ascending=True, inplace=True)  # ascending=True : 오름차순(기본값)
df=df.sort_values(['나이'], ascending=False) # ascending=False : 내림차순
df
```

```python
# 상위 3개의 데이터 출력
display(df.loc[0:4])
display(df.iloc[0:3])
```

```python
# index 값 기준 오름차순 정렬
df=df.sort_index()   # ascending=True
df
```

##### 행/열 데이터 추가

```python
# 행 추가
df.loc[5]=[6, "FFF", "여", "26"]   # 마지막 index값 + 1
df
```

```python
df.loc[5]=[7, "GGG", "여", "28"]
df
```

```python
# 열 추가
df['점수'] = [80, 88, 76, 92, 68, 55]
df
```

```python
# 열(필드)이름 변경1: 점수 => 국어 변경
df=df.rename(columns={"점수":"국어"})
df
```

```python
# 열(필드)이름 변경1: 여러개를 동시에 변경하는 방법
df['점수'] = [92,86,75,48,93,79]
df.columns=['번호', '이름', '성별', '나이', '국어', '수학']
df
```

##### DataFrame 연산

```python
df['합계'] = df['국어'] + df['수학']
df
```

```python
# 데이터 정보 확인
df.dtypes  # 데이터 타입 확인
df.info()  # 데이터 전체 정보 확인
```

```python
# 데이터 타입 변경
df=df.astype({"나이":int, '국어':int})
df.info()
```

```python
df['만나이'] = df['나이'] - 1
df
```
