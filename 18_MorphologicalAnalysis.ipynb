#### 영문 형태소 분석

```python
from nltk.tokenize import RegexpTokenizer # 정규식 표현에 의한 데이터 전처리 함수
from nltk import Text
import matplotlib.pyplot as plt
```

```python
text='''
n land And now I understand What you tried to say to me How you suffered for yo
you suffered for your sanity How you tried to set them free They would not list
g hand And now I understand What you tried to say to me How you suffered for yo
you suffered for your sanity How you tried to set them free They would not list
snow And now I think I know What you tried to say to me How you suffered for yo
you suffered for your sanity How you tried to set them free They would not list
'''
```

```python
regTok = RegexpTokenizer("[\w]+") # [0-9a-zA-Z ], 영문/숫자를 제외한 나머지 글자 제거
tokText=Text(regTok.tokenize(text))
print(tokText)
print(type(tokText))
```

```python
# plt.figure(figsize=(10, 5))  # 프로젝트에서 1회성 지정(plt.show() 실행시 값 사라짐)
plt.rcParams['figure.figsize']=(10, 5)  # 계속해서 사용 가능
plt.title('NLTK Test')
tokText.plot()
plt.show()
```

```python
tokText.dispersion_plot(['you', 'to', 'for', 'free'])
```

```python
# for와 if 한줄쓰기
nouns_list=[t[0] for t in tag_list if t[1] == 'NNP' or t[1] == 'NN']
print(nouns_list)
```

```python
# 단어 빈도수 계산하기
import pandas as pd

tokser=pd.Series(nouns_list)

print(tokser.value_counts()) # 단어 빈도수 계산
```

#### 2023년 대통령 신년사 형태소 분석

```python
# 모듈 연결
from konlpy.tag import Okt, Mecab
from nltk import Text
from collections import Counter
from wordcloud import WordCloud
import nltk
```

```python
f = open('/content/drive/MyDrive/Notability/Github/Python_Machine_Learning/2023년 대통령 신년사.txt','r')
text = f.read()
text=text.replace('\n\n','\n').replace('\n',' ')
text
```

```python
# Okt, Mecab 인스턴스 생성 및 품사단위 형태소 나누기
okt=Okt()
mecab=Mecab()

start=time.time()
okt_pos=okt.pos(text) # Okt를 품사 단위 분류
print(f'okt => {time.time()-start:.4f} sec')

start=time.time()
mecab_pos=mecab.pos(text) # mecab를 품사 단위 분류
print(f'mecab => {time.time()-start:.4f} sec')
```

```python
print(okt_pos)
print(mecab_pos)
```

```python
# Okt 결과에서 명사(Noun), 형용사(Adjective), 부사(Verb) 추출
Nonus= [t[0] for t in okt_pos if t[1] == 'Noun']
Adjs= [t[0] for t in okt_pos if t[1] == 'Adjective'] 
Verbs= [t[0] for t in okt_pos if t[1] == 'Verb'] 

print(Nonus[:5])
print(Adjs[:5])
print(Verbs[:5])
```

```python
# Mecab 결과에서 명사(Noun), 형용사(Adjective), 부사(Verb) 추출
Nonus= [t[0] for t in mecab_pos if t[1] == 'NNG']
Adjs= [t[0] for t in mecab_pos if t[1] == 'MAG'] 
Verbs= [t[0] for t in mecab_pos if t[1] == 'VV'] 

print(Nonus[:5])
print(Adjs[:5])
print(Verbs[:5])
```

```python
# 데이터 하나로 합치기
morph = Nonus + Adjs + Verbs
len(morph)
```

```python
# nltk.Text() 함수를 이용해 데이터 탐색 진행한다.
morph[:10]
morphT = Text(morph)
morphT
```

```python
import matplotlib.font_manager as fm

# path='NanumGothic.ttf'
# font_name=fm.FontProperties(fname=path, size=10).get_name()

plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(10, 3))

plt.title('2023년 대통령 신년사')
morphT.plot()
plt.show()
```

```python
# 빈도수 계산
import pandas as pd

morph_counts=pd.Series(morph).value_counts()
morph_df=morph_counts.reset_index()
morph_df.columns=['단어','빈도수']
morph_df.head(15)
```

```python
plt.figure(figsize=(10, 3))

morphT.dispersion_plot(['국민', '경제', '세계','개혁','기술','교육'])
plt.show()
```

```python
from collections import Counter

morph_cnt = Counter(morph) # 리스트에서 입력된 데이터 빈도수 계산
morph_cnt                   # 결과는 딕셔너리 구조로 출력
```

```python
# 딕셔너리 자료 구조를 튜플구조로 변경(키와 값이 1대1일 경우에만 가능)
morph_txt=morph_cnt.most_common()
print(morph_txt[:10])
```

```python
# 빈도수가 낮은 데이터는 제거(3 미만인 데이터 제거)
morph_txt_up = [tu for tu in morph_txt if tu[-1] >= 3]
print(len(morph_txt_up))
print(morph_txt_up[:5])
```

```python
# 변경한 튜플형 데이터를 다시 딕셔너리 형으로 변경
morph_txt_dict = dict(morph_txt_up)
morph_txt_dict
```

```python
# 워드클라우드 객체 생성
wordcloud=WordCloud(font_path='NanumBarunGothic',
                    background_color='black',
                    width=1000,height=1000,
                    max_words=100,max_font_size=300)

# 워드클라우드 출력
wc=wordcloud.generate_from_frequencies(morph_txt_dict)
plt.figure(figsize=(8, 8))
plt.imshow(wc)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### 불필요한 단어(불용어(Stopword)) 제거


```python
stop_words='새로운 세계 나라 모든 수출 되는 하겠습니다 없습니다 입니다 합니다 있습니다 하는 우리 통해 경제 수 그 것 더 역시 할 대한 국민 해 여러분 있도록 주의 위해'
stop_words=stop_words.split()
stop_words
```

```python
morph_stop_word=[ t for t in morph if t not in stop_words]
# not in -> 포함되어 있지 않다면 t에 차곡차곡 쌓아라.
print(len(morph))
print(len(morph_stop_word))
```

```python
## 단어 빈도수 계산
morph_cnt = Counter(morph_stop_word)

# 튜블 구조 생성
morph_txt=morph_cnt.most_common()

# 빈도수가 3 이상인 데이터만 추출
morph_txt_up = [ tu for tu in  morph_txt if tu[-1] >= 3 ]

# 딕셔너리 구조로 변경
morph_txt_dict = dict(morph_txt_up)

# 워드클라우드 출력
wc=wordcloud.generate_from_frequencies(morph_txt_dict)
plt.figure(figsize=(8, 8))
plt.imshow(wc)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
