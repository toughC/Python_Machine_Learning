```python
### python 함수정의: 특정한 행동만 진행하는 함수
def prnMax(a, b):
    if a > b:
        print(f'{a}가 {b}보다 큰 값입니다.')
    elif a == b:
        print(f'{a}와 {b}는 동일 값입니다.')
    else:
        print(f'{b}가 {a}보다 큰 값입니다.')
```

```python
### 함수 호출
x=10; y=10
prnMax(x, y)
```

```python
### python 함수정의: 변수를 이용해 계산후 결과를 돌려주는 함수
def add(a, b=0):
    tot=a+b

    if a>b:pMax=a
    elif a<b:pMax=b
    else:pMax="같다"

    return tot, pMax
```

```python
aa=10;bb=20
print(add(aa))
print(add(-aa))
print(add(aa, bb))
t, m = add(10, 20)
print(t, m)
```
