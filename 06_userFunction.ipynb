```python
### python 함수정의: 특정한 행동만 진행하는 함수
def prnMax(a, b):
    if a > b:
        print(f'{a}가 {b}보다 큰 값입니다.')
    elif a == b:
        print(f'{a}와 {b}는 동일 값입니다.')
    else:
        print(f'{b}가 {a}보다 큰 값입니다.')
```

```python
### 함수 호출
x=10; y=10
prnMax(x, y)
```

```python
### python 함수정의: 변수를 이용해 계산후 결과를 돌려주는 함수
def add(a, b=0):
    global aa  # 전역변수 선언
    aa=20

    tot=a+b
    if a>b:
        pMax=a   # pMax 지역변수(함수에서 정의된 변수 => 지역변수)
    elif a<b:
        pMax=b
    else:
        pMax="같다"
    return tot, pMax
```

```python
aa=10;bb=20 # 전역변수

print(add(aa))
print(add(-aa))
print(add(aa, bb))
t, m = add(aa, bb)
print(t, m)
print(aa)
```

```python
# 클래스 정의

class 빵틀:
    def 빵반죽(self):  # 메서드(멤버함수) 
        self.기본재료='밀가루'   # 애트리뷰트 정의  
        self.필수추가='이스트'   # 애트리뷰트 정의
        self.기본시간=5          # 애트리뷰트 정의

    def 빵모양(self): # 메서드(멤버함수)
        self.크기가로=20        # 애트리뷰트 정의
        self.크기세로=5         # 애트리뷰트 정의
```

```python
# 빵틀을 이용해 객체 생성
밤빵=빵틀()   # 객체(인스턴스) 생성
밤빵.추가재료='밤'  # 클래스에 없는 객체 애트리뷰트(속성) 생성 및 값 설정
밤빵.칼로리=100     # 클래스에 없는 객체 애트리뷰트(속성) 생성 및 값 설정

버터빵=빵틀()  # 객체(인스턴스) 생성
버터빵.추가재료='버터'  # 클래스에 없는 객체 애트리뷰트 생성 및 값 설정
버터빵.칼로리=200       # 클래스에 없는 객체 애트리뷰트 생성 및 값 설정
```

```python
밤빵.빵반죽()
밤빵.빵모양()

print(f'{밤빵.기본재료}/{밤빵.필수추가}/{밤빵.기본시간}/{밤빵.추가재료}/{밤빵.칼로리}/{밤빵.크기가로}/{밤빵.크기세로}')

밤빵.기본재료="밀가루+밤"
print(f'{밤빵.기본재료}/{밤빵.필수추가}/{밤빵.기본시간}/{밤빵.추가재료}/{밤빵.칼로리}/{밤빵.크기가로}/{밤빵.크기세로}')
```
