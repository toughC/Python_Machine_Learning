```python
import pandas as pd

### list 데이터를 이용한 DataFrame 생성
lstdata=[[1, "홍길동", "남", 500],
         [2, "김파이", "남", 35],
         [3, '박데이', "여", 55],
         [4, '이분석', '여', 60],
         [5, '박판다', '남', 30]
         ]
         
# 행/열 인덱스를 기본값으로 부여(숫자)
df=pd.DataFrame(lstdata)  
df
```

```python
# 열 인덱스를 사용자가 원하는 이름으로 부여  
df=pd.DataFrame(lstdata, 
                columns=['번호', '이름', '성별', '나이'])  
df
```

```python
### list 데이터를 이용한 DataFrame 생성
lstdata=[[1, "홍길동", "남", ],
         [2, "김파이", "남", 35],
         [3, '박데이', "여", 55],
         [4, '이분석', '여', 60],
         [5, '박판다', '남', 30]
         ]
df=pd.DataFrame(lstdata)
df
```

#### pandas 행/열 데이터(원소) 출력 

```python
# 딕셔너리 데이터를 이용한 DataFrame 구성
dictdata={"번호":[1,2,3,4,5],
          "이름":["AAA", "BBB", "CCC", "DDD", "EEE"],
          "성별":["남", "남", "여", "여", "남"],
          "나이":[50,40,30,35,40]
          }

df=pd.DataFrame(dictdata)
df
```

```python
# 행 index를 이용한 데이터 출력(기본)
print(df.loc[2])   # index가 2인 데이터 출력
display(df.loc[1:3])   # index가 1 ~ 3까지의 데이터 출력
```

```python
# 열 index를 이용한 데이터 출력
display(df['번호'])   # 1개의 열 출력
display(df[['번호', '이름']])   # 1개의 열 출력
display(df.loc[:, "이름"])   # 1개의 열 출력
display(df.loc[:, ["이름", '번호']]) 
```

```python
display(df.loc[1:3, ["이름", '번호']]) 
```

#### 행/열 추가 및 삭제

```python
# DataFrame() 함수로 데이터프레임 변환. 변수 df에 저장 
exam_data = {'수학' : [ 90, 80, 70], 
             '영어' : [ 98, 89, 95],
             '음악' : [ 85, 95, 100], 
             '체육' : [ 100, 90, 90]}

df = pd.DataFrame(exam_data, index=['서준', '우현', '인아'])
df
```

```python
# DataFrame 복사
df1 = df.copy()
# 행 기준 데이터 삭제
# df1=df1.drop(["우현"])   # 우현 데이터 삭제후 df1 데이터 업데이트
df1.drop(["우현"], inplace=True) # 우현 데이터 삭제후 즉시 적용
```

```python
df1 = df.copy()
df1.drop(["우현"], axis=0, inplace=True) # axis=0 : 행 기준, axis=1 : 열기준
df1
```

```python
df1 = df.copy()
df1.drop(["우현",'서준'], axis=0, inplace=True) # 두개 이상의 행 삭제
df1
```

#### 열 제거

```python
df1 = df.copy()
df1.drop(['수학', '체육'], axis=1, inplace=True) # 두개의 열 삭제
df1
```

```python
df1 = df.copy()
df1 = df1.drop(['수학', '체육'], axis=1) # 두개의 열 삭제
df1
```
