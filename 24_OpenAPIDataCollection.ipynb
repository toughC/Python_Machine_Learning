### API란?
- Application Programming Interface (애플리케이션 프로그래밍 인터페이스)
- 애플리케이션(스마트폰, 카카오톡 서버 등 서로 다른 프로그램)이 인터페이싱하는(요청과 응답을 주고받는) 체계"라고 이해

#### Open API
- "누구나 쓸 수 있도록 공개된(Open) API"
- 대표적으로, 정부의 공공데이터포털(www.data.go.kr)로 해당 사이트를 통해 국가기관이 보유한 수많은 데이터를 API 형태로 무료 공개
- 카카오, 네이버, Google 등 수많은 민간 기업들 또한 자사가 제공하는 다양한 기능들(지도 앱, 포털 검색하기, 로그인 기능 등)을 일반 사람들도 무료로 사용할 수 있도록 API 형태로 제공

#### 부산명소정보 전체 받아오기

```python
from bs4 import BeautifulSoup as bs  
import requests  
import pandas as pd
import json
```

```python
user_key='VFYXI%2F3A5kIk7QVsO1epAlZSb6ZtCO9N5VC%2FMF3pW6oZ7%2BsFrvH1Y9y0UfmVzTnA0k2FasoNM6t6JsY7%2BEwNEw%3D%3D'
totRow=10

url='http://apis.data.go.kr/6260000/AttractionService/getAttractionKr?serviceKey='+user_key
url += '&resultType=json&numOfRows='+str(totRow)+'&pageNo=1'

bu_html=requests.get(url)
if bu_html.status_code != 200: exit('데이터를 받지 못했습니다.')

bu_json=json.loads(bu_html.text)
totRow=bu_json['getAttractionKr']['totalCount']
totRow
```

```python
url='http://apis.data.go.kr/6260000/AttractionService/getAttractionKr?serviceKey='+user_key
url += '&resultType=json&numOfRows='+str(totRow)+'&pageNo=1'

bu_html=requests.get(url)
if bu_html.status_code != 200: exit('데이터를 받지 못했습니다.')

bu_json=json.loads(bu_html.text)

df=pd.DataFrame(bu_json['getAttractionKr']['item'])
df.info()
```

```python
df_sub=df[['MAIN_TITLE', 'LAT', 'LNG','ADDR1']].copy()
df_sub.head()
```

```python
df_sub['sigu']=df_sub['ADDR1'].str.split(' ').str[1]
df_sub.head()
```

```python
df_sub.isna().sum()
```

```python
# sigu NaN 값 변경
gu_lst=['연제구', '중구', '수영구']
for idx, i in enumerate(df_sub[df_sub['sigu'].isna()].index):
    df_sub.loc[i, 'sigu'] = gu_lst[idx]

df_sub.isna().sum()
```

```python
# 부산 구/군 단위 명소 통계
sigu_cnt=df_sub.groupby(by='sigu')[['MAIN_TITLE']].count()
sigu_cnt.sort_values('MAIN_TITLE', ascending=False, inplace=True)

#plt.rc('font', family='NanumBarunGothic')
sigu_cnt.plot(kind='barh')
```
